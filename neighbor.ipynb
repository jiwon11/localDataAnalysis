```python
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point, Polygon, LineString
import matplotlib
import pandas as pd
from matplotlib_scalebar.scalebar import ScaleBar
```

```python
print(gpd.__version__)
```

```python
area = gpd.GeoDataFrame.from_file('./EMD_202005/EMD.shp', encoding='cp949')
seoul_area = area[area['EMD_CD'].str.startswith('11')==True]  
```

```python
seoul_area.plot(column='EMD_ENG_NM')
```

```python
gwanggyo_area = area[((area['EMD_CD'].str.startswith('411')==True) | (area['EMD_CD'].str.startswith('414')==True)) & ((area['EMD_KOR_NM'] == '이의동') | (area['EMD_KOR_NM'] == '원천동') | (area['EMD_KOR_NM'] == '하동') | (area['EMD_KOR_NM'] == '상현동'))]
gwanggyo_area
```

```python
gwanggyo_area.plot(column='EMD_ENG_NM')
```

```python
gyeonggi_dentist = pd.read_excel('./dentist/gg_dentist.xls',encoding='utf-8', dtype=str)
gyeonggi_dentist['경도'] = gyeonggi_dentist['WGS84경도'].astype(float)
gyeonggi_dentist['위도'] = gyeonggi_dentist['WGS84위도'].astype(float)
gyeonggi_dentist['geometry'] = gyeonggi_dentist.apply(lambda row : Point([row['경도'], row['위도']]), axis=1)
gyeonggi_dentist = gpd.GeoDataFrame(gyeonggi_dentist, geometry='geometry')
gyeonggi_dentist.crs = {'init':'epsg:4326'}
gyeonggi_dentist = gyeonggi_dentist.to_crs({'init':'epsg:5179'})
```

```python
gyeonggi_dentist
```

```python
#select_dt = gyeonggi_dentist.within(gwanggyo_area['geometry'])
gw_wcd_dentist = gyeonggi_dentist[(gyeonggi_dentist.within(gwanggyo_area.loc[:, 'geometry'].iloc[0])==True)]
gw_iud_dentist = gyeonggi_dentist[gyeonggi_dentist.within(gwanggyo_area.loc[:, 'geometry'].iloc[1])==True]
gw_had_dentist = gyeonggi_dentist[gyeonggi_dentist.within(gwanggyo_area.loc[:, 'geometry'].iloc[2])==True]
gw_shd_dentist = gyeonggi_dentist[gyeonggi_dentist.within(gwanggyo_area.loc[:, 'geometry'].iloc[3])==True]
```

```python
print(type(gw_shd_dentist))
```

```python
gw_dentists = pd.concat([gw_wcd_dentist,gw_iud_dentist,gw_had_dentist,gw_shd_dentist],axis=0)
gw_dentists = gw_dentists[gw_dentists['영업상태명']=='영업중']
gw_dentists.info()
```

```python
wgw_data = {'lat': [127.0432819],
            'long': [37.2942124]
           }

wgw = pd.DataFrame(wgw_data)
wgw['geometry'] = wgw.apply(lambda row : Point([row['lat'], row['long']]), axis=1)
wgw = gpd.GeoDataFrame(wgw, geometry=wgw['geometry'])
wgw.crs = {'init':'epsg:4326'}
wgw = wgw.to_crs({'init':'epsg:5179'})
wgw

gwanggyo_area = gwanggyo_area.dissolve(by='EMD_CD')
ax = gwanggyo_area.plot(column="EMD_ENG_NM",categorical=True,legend=True, figsize=(15,15), alpha=0.8)
gw_dentists.plot(ax=ax, marker='o', color='green', label='dentist')
wgw.plot(ax=ax, marker='v', color='black')
ax.set_title("Gwanggyo", fontsize=20)
ax.set_axis_off()
scalebar = ScaleBar(0.2) # 1 pixel = 0.2 meter
scalebar.location = 'upper left'
plt.gca().add_artist(scalebar)
plt.show()
```
